### 1.  Scrape the details of most viewed videos on YouTube from Wikipedia: 
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/ 

You need to find following details: 

A) Rank 

B) Name 

C) Artist 

D) Upload date 

E) Views

```python
url =  'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver =  webdriver.Chrome("chromedriver.exe")
driver.get(url)
```

```python
num =[]
vid =[]
uploadr = []
views = []
date = []
```

```python
for i in range(1,31):
    num_tag=driver.find_element_by_xpath(f"//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr[{i}]/td[1]")
    try:
      num.append(num_tag.text)
    except NoSuchElementException:
      num.append("-")
       
for i in range(1,31):
    vid_tag=driver.find_element_by_xpath(f"//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr[{i}]/td[2]")
    try:
        vid.append(vid_tag.text.strip().split('"')[1]) #appending only the names of the videos
    except NoSuchElementException:
        vid.append("-")
 
for i in range(1,31):
    uploadr_tag=driver.find_element_by_xpath(f"//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr[{i}]/td[3]")
    try:
        uploadr.append(uploadr_tag.text)
    except NoSuchElementException:
        uploadr.append("-")
        
for i in range(1,31):
    views_tag=driver.find_element_by_xpath(f"//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr[{i}]/td[4]")
    try:
        views.append(views_tag.text)
    except NoSuchElementException:
        views.append("-")
        
        
for i in range(1,31):
    date_tag=driver.find_element_by_xpath(f"//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr[{i}]/td[5]")
    try:
        date.append(date_tag.text)
    except NoSuchElementException:
        date.append("-")

topyoutubevids = {"Rank":num,"Name":vid,"Artist":uploadr,"Upload Date":date,"Views(Billions)":views}
topyoutubevidsDF = pd.DataFrame(topyoutubevids)
```
